```python
import numpy as np
def levenshtein(seq1, seq2):
    size_x = len(seq1) + 1
    size_y = len(seq2) + 1
    matrix = np.zeros ((size_x, size_y))
    for x in range(size_x):
        matrix [x, 0] = x
    for y in range(size_y):
        matrix [0, y] = y

    for x in range(1, size_x):
        for y in range(1, size_y):
            if seq1[x-1] == seq2[y-1]:
                matrix [x,y] = min(
                    matrix[x-1, y] + 1,
                    matrix[x-1, y-1],
                    matrix[x, y-1] + 1
                )
            else:
                matrix [x,y] = min(
                    matrix[x-1,y] + 1,
                    matrix[x-1,y-1] + 1,
                    matrix[x,y-1] + 1
                )
    score = matrix[size_x - 1, size_y - 1]
#     final_score = 100-(score/len(seq1)*100)
    return score
```

### Implementing on small testset to chek the validity

```python
total = []
collect =[]
for ind,countries in  test_run.iterrows():
    current_cities  = countries['misspelt_name']
    seq2 = current_cities.lower()
    country_to_search = countries['country']
    rslt_df = correct_cities[correct_cities['country'] == country_to_search]
    min_score =100
    for index,row in rslt_df.iterrows():
        seq1 =  row['name'].lower()
        if len(seq1) == len(seq2):
            score = levenshtein(seq2, seq1)
            if score < min_score:
                collect  = [score,seq2,seq1,row['id']]
                min_score = score
    total.append(collect)
print(total)
```
